```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

src = cv2.imread('HW1-2  Fig0525(b)(aerial_view_turb_c_0pt0025) .tif', cv2.IMREAD_GRAYSCALE)
height, width = src.shape[0], src.shape[1]

plt.hist(src.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.show()

histogram = np.zeros(256)
lookUpTable = np.zeros(256)
    
for i in range(height):
    for j in range(width):
        histogram[src[i][j]] += 1

sum = 0.0
scale_factor = 255.0/(height * width)

for i in range(256):
    sum += histogram[i]
    lookUpTable[i] = round( sum * scale_factor )

for i in range(height):
    for j in range(width):
        src[i][j] = lookUpTable[src[i][j]]

plt.hist(src.flatten(),256,[0,256], color = 'b')
plt.xlim([0,256])
plt.show()

cv2.imshow("src",src)
cv2.waitKey()
cv2.destroyAllWindows()
```
