```python
import cv2
import numpy as np
```

```python
# 피부색 추출
vidcap = cv2.VideoCapture('./example/Project- hand gesture.AVI')
count = 0 
while(vidcap.isOpened()): 
    ret, image = vidcap.read() 
    if(int(vidcap.get(1)) % 50 == 0): 
        print('Saved frame number : ' + str(int(vidcap.get(1)))) 
        image = cv2.GaussianBlur(image, (5,5),0)
        # BGR -> YCrCb 변환
        YCrCb = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)
        # 피부 검출
        mask_hand = cv2.inRange(YCrCb,np.array([0,133,77]),np.array([255,173,127]))
        # 피부 색 나오도록 연산
        mask_color = cv2.bitwise_and(image,image, mask=mask_hand)
        # 추출된 이미지가 저장되는 경로 
        cv2.imwrite("./result/frame%d.png" % count, mask_color) 
        #print('Saved frame%d.jpg' % count) 
        count += 1 
        
vidcap.release()
```
